```python
import numpy as np
import pandas as pd
arreglos= np.array([["Bella", "Labrador","Café",56,24,"2013-07-01"],["Charlie", "Poodle","Negro",56,24,"2016_09_16"],["Lucy", "Chow Chow","Café",46,24,"2014-08-25"],["Cooper", "Schnauzer","Gris",49,17,"2011-12-11"],["Bernie", "San Bernardo","Blanco",77,74,"2018-02-27"],["Max", "Chow Chow","Blanco",45,28,"2019-03-27"],["Red", "Schnauzer","Negro",45,15,"2017-09-18"]])
arreglos
```

```python
df= pd.DataFrame(data=arreglos,columns=["Nombre","Raza","Color","Altura","Peso","Nacimiento"])
df["Altura"]= df["Altura"].astype(int)
df["Peso"]= df["Peso"].astype(int)
df
```

```python
# Extracción con tabla pivote

df_altura_por_raza_color= df.pivot_table( "Altura", index="Raza", columns="Color")
df_altura_por_raza_color
```

```python
# EXTRACCIÓN DE REGISTROS TABLA PIVOTE

df_altura_por_raza_color.loc["Labrador":"San Bernardo"]
```

```python
# Cálculo de promedio para todos los renglones (por columnas)

df_altura_por_raza_color.mean(axis="index")
```

```python
# Cálculo de promedio para todos las columnas (por renglón)

df_altura_por_raza_color.mean(axis="columns")
```
